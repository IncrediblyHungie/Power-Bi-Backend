```python
import pandas as pd
from datetime import datetime

# Assuming cashpro and bluefin are DataFrames with relevant columns

# For Cashpro
# Convert Report Date to datetime and filter for last month
cashpro['Report Date'] = pd.to_datetime(cashpro['Report Date'])
last_month = cashpro['Report Date'].dt.month.max()  # Assuming the data is up to date and last month refers to the last month in the dataset
cashpro_last_month = cashpro[cashpro['Report Date'].dt.month == last_month]

# Find top 3 most common values for Rtn/NOC Code
cashpro_top3_codes = cashpro_last_month['Rtn/NOC Code'].value_counts().nlargest(3)
cashpro_codes_percent = (cashpro_top3_codes / cashpro_last_month['Rtn/NOC Code'].count()).round(2)
# Sum and Average of Amount for Cashpro last month
cashpro_last_month['Amount'] = cashpro_last_month['Amount'].replace('[\\$,]', '', regex=True).astype(float)
cashpro_last_month_sum = cashpro_last_month['Amount'].sum()
cashpro_last_month_avg = cashpro_last_month['Amount'].sum() / cashpro_last_month['Report Date'].nunique()
print("Cashpro - Top 3 Rtn/NOC Codes for Last Month and their percentage:")
print(cashpro_codes_percent)
cashpro_to_csv = pd.DataFrame({'Code': cashpro_codes_percent.index, 'Percentage': cashpro_codes_percent.values, 'Amount': [cashpro_last_month_sum, None, None], 'Monthly Average': [None, cashpro_last_month_avg, None]})
cashpro_to_csv.to_csv('C:\\Users\\pskotte\\Desktop\\Power Bi Data\\DM Monthly Average.csv', index=False)

# For ILMS
# Duplicate process for ILMS
ilms['Report Date'] = pd.to_datetime(ilms['Report Date'])
ilms_last_month = ilms[ilms['Report Date'].dt.month == last_month]

ilms_top3_codes = ilms_last_month['Rtn/NOC Code'].value_counts().nlargest(3)
ilms_codes_percent = (ilms_top3_codes / ilms_last_month['Rtn/NOC Code'].count()).round(2)

ilms_last_month['Amount'] = ilms_last_month['Amount'].replace('[\\$,]', '', regex=True).astype(float)
ilms_last_month_sum = ilms_last_month['Amount'].sum()
ilms_last_month_avg = ilms_last_month['Amount'].sum() / ilms_last_month['Report Date'].nunique()
print("ILMS - Top 3 Rtn/NOC Codes for Last Month and their percentage:")
print(ilms_codes_percent)

ilms_to_csv = pd.DataFrame({'Code': ilms_codes_percent.index, 'Percentage': ilms_codes_percent.values, 'Amount': [ilms_last_month_sum, None, None], 'Monthly Average': [None, ilms_last_month_avg, None]})
ilms_to_csv.to_csv('C:\\Users\\pskotte\\Desktop\\Power Bi Data\\ILMS Monthly Average.csv', index=False)

# For Bluefin
# Convert Transaction Date Time to datetime and filter for last month
bluefin['Transaction Date Time'] = pd.to_datetime(bluefin['Transaction Date Time'])
bluefin_last_month = bluefin[bluefin['Transaction Date Time'].dt.month == last_month]

# Approved Amount
approved_amount_monthly_sum = bluefin_last_month[bluefin_last_month['Payment Processor Response Code Description'] == 'Transaction Approved']['Amount'].replace('[\\$,]', '', regex=True).astype(float).sum()
approved_amount_monthly_avg = approved_amount_monthly_sum / bluefin_last_month['Transaction Date Time'].nunique()

# Exclude 'Approved' and find top 3 values for Payment Processor Response Code Description excluding 'Approved'
bluefin_last_month_excluding_approved = bluefin_last_month[bluefin_last_month['Payment Processor Response Code Description'] != 'Approved']
bluefin_top3_responses_excluding_approved = bluefin_last_month_excluding_approved['Payment Processor Response Code Description'].value_counts().nlargest(3)
bluefin_responses_percent = (bluefin_top3_responses_excluding_approved / bluefin_last_month_excluding_approved['Payment Processor Response Code Description'].count()).round(2)
# Sum and Average of Amount for Bluefin last month excluding 'Approved'
bluefin_last_month_excluding_approved['Declined Amount'] = bluefin_last_month_excluding_approved['Amount'].replace('[\\$,]', '', regex=True).astype(float)
bluefin_last_month_excluding_approved_sum = bluefin_last_month_excluding_approved['Declined Amount'].sum() / bluefin_last_month['Transaction Date Time'].dt.day.nunique()
print("\\nBluefin - Top 3 Payment Processor Response Code Descriptions for Last Month (excluding 'Approved') and their percentage:")
print(bluefin_responses_percent)
bluefin_to_csv = pd.DataFrame({'Response Code Description': bluefin_responses_percent.index, 'Percentage': bluefin_responses_percent.values, 'Average Declined Amount': [bluefin_last_month_excluding_approved_sum, None, None], 'Approved Amount': [approved_amount_monthly_sum, approved_amount_monthly_avg, None]})
bluefin_to_csv.to_csv('C:\\Users\\pskotte\\Desktop\\Power Bi Data\\Card Monthly Average.csv', index=False)

# Assuming cashpro_daily and bluefin_daily are DataFrames with relevant columns
cashpro_daily = cashpro
# For Cashpro daily values
# Convert Report Date to datetime and filter for today
cashpro_daily['Report Date'] = pd.to_datetime(cashpro_daily['Report Date'])
today = datetime.now().date()  # Getting today's date
cashpro_daily_today = cashpro_daily[cashpro_daily['Report Date'].dt.date == today]

# Find top 3 most common values for Rtn/NOC Code
cashpro_daily_top3_codes = cashpro_daily_today['Rtn/NOC Code'].value_counts().nlargest(3)
cashpro_daily_codes_percent = (cashpro_daily_top3_codes / cashpro_daily_today['Rtn/NOC Code'].count()).round(2)
# Sum of Amount for Cashpro today
cashpro_daily_today['Amount'] = cashpro_daily_today['Amount'].replace('[\\$,]', '', regex=True).astype(float)
cashpro_daily_sum = cashpro_daily_today['Amount'].sum()
print("Cashpro daily - Top 3 Rtn/NOC Codes for Today and their percentage:")
print(cashpro_daily_codes_percent)
cashpro_daily_to_csv = pd.DataFrame({'Code': cashpro_daily_codes_percent.index, 'Percentage': cashpro_daily_codes_percent.values, 'Amount': [cashpro_daily_sum, None, None]})
cashpro_daily_to_csv.to_csv('C:\\Users\\pskotte\\Desktop\\Power Bi Data\\DM Today.csv', index=False)

# For ILMS daily values, exact same process as cashpro
ilms_daily = ilms
ilms_daily['Report Date'] = pd.to_datetime(ilms_daily['Report Date'])
ilms_daily_today = ilms_daily[ilms_daily['Report Date'].dt.date == today]
ilms_daily_top3_codes = ilms_daily_today['Rtn/NOC Code'].value_counts().nlargest(3)
ilms_daily_codes_percent = (ilms_daily_top3_codes / ilms_daily_today['Rtn/NOC Code'].count()).round(2)
ilms_daily_today['Amount'] = ilms_daily_today['Amount'].replace('[\\$,]', '', regex=True).astype(float)
ilms_daily_sum = ilms_daily_today['Amount'].sum()
print("ILMS daily - Top 3 Rtn/NOC Codes for Today and their percentage:")
print(ilms_daily_codes_percent)
ilms_daily_to_csv = pd.DataFrame({'Code': ilms_daily_codes_percent.index, 'Percentage': ilms_daily_codes_percent.values, 'Amount': [ilms_daily_sum, None, None]})
ilms_daily_to_csv.to_csv('C:\\Users\\pskotte\\Desktop\\Power Bi Data\\ILMS Today.csv', index=False)

# For Bluefin daily values
# Convert Transaction Date Time to datetime and filter for today
bluefin_daily = bluefin
bluefin_daily['Transaction Date Time'] = pd.to_datetime(bluefin_daily['Transaction Date Time'])
bluefin_daily_today = bluefin_daily[bluefin_daily['Transaction Date Time'].dt.date == today]

approved_amount_daily_sum = bluefin_daily_today[bluefin_daily_today['Payment Processor Response Code Description'] == 'Transaction Approved']['Amount'].replace('[\\$,]', '', regex=True).astype(float).sum()

# Exclude 'Approved' and find top 3 values for Payment Processor Response Code Description excluding 'Approved'
bluefin_daily_today_excluding_approved = bluefin_daily_today[bluefin_daily_today['Payment Processor Response Code Description'] != 'Approved']
bluefin_daily_top3_responses_excluding_approved = bluefin_daily_today_excluding_approved['Payment Processor Response Code Description'].value_counts().nlargest(3)
bluefin_daily_responses_percent = (bluefin_daily_top3_responses_excluding_approved / bluefin_daily_today_excluding_approved['Payment Processor Response Code Description'].count()).round(2)
# Sum of Amount for Bluefin today excluding 'Approved'
bluefin_daily_today_excluding_approved['Declined Amount'] = bluefin_daily_today_excluding_approved['Amount'].replace('[\\$,]', '', regex=True).astype(float)
bluefin_daily_excluding_approved_sum = bluefin_daily_today_excluding_approved['Declined Amount'].sum()
print("\\nBluefin daily - Top 3 Payment Processor Response Code Descriptions for Today (excluding 'Approved') and their percentage:")
print(bluefin_daily_responses_percent)
bluefin_daily_to_csv = pd.DataFrame({'Response Code Description': bluefin_daily_responses_percent.index, 'Percentage': bluefin_daily_responses_percent.values, 'Declined Amount': [bluefin_daily_excluding_approved_sum, None, None], 'Approved Amount': [approved_amount_daily_sum, None, None]})
bluefin_daily_to_csv.to_csv('C:\\Users\\pskotte\\Desktop\\Power Bi Data\\Card Processing Today.csv', index=False)
```
